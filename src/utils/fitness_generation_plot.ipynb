```python
import json
import matplotlib.pyplot as plt

# Load data from the JSON file
with open('../result/norm_cp_2/test_suite_evolution.json', 'r') as file: # modify here for your own use
    lines = file.readlines()

# Extract x and y values for plotting
x_values = []
y_values = []

for line in lines:
    data = json.loads(line)
    key, value = data.popitem()
    generation = value['generation']
    fitness = value['fitness']

    # Extract the x value from the key
    x_value = int(key.split('_')[1])

    x_values.append(x_value)
    y_values.append(fitness)

# Plotting the graph
plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.title('Fitness Evolution Over Time (Norm CP select)') # modify here for your own use
plt.xlabel('Generation')
plt.ylabel('Fitness Value')
plt.grid(True)
plt.show()
```
